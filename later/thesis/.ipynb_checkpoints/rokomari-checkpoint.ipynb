```python
## importing the parsing module
from bs4 import BeautifulSoup
import requests
from difflib import get_close_matches
```

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246'
headers = {'User-Agent': user_agent}
key =input()
url="https://www.rokomari.com/search?term="+str(key)
```

```python
url
```

```python
r = requests.get(url)
r.text
```

```python
soup = BeautifulSoup(r.content, 'lxml')
```

```python
soup
```

```python
name=[]
for node in soup.findAll('div', attrs={'class':'browse__content-books-wrapper'}):
    for item in (node.findAll('p',attrs={'class':'book-title'})):
        name.append(item.text)
```

```python
name
```

```python
result=[]
for node in soup.findAll('div', attrs={'class':'browse__content-books-wrapper'}):
    for item in (node.findAll('p',attrs={'class':'book-price'})):
        result.append(item.text.strip())
```

```python
result
```

```python
d={}
for x,y in zip(name,result):
    d.update({x:y})
```
