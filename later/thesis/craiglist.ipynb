```python
## importing the parsing module
from bs4 import BeautifulSoup
import requests
from difflib import get_close_matches
```

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246'
headers = {'User-Agent': user_agent}
key =input()
url='https://newjersey.craigslist.org/search/sss?query='+str(key)+'&sort=rel'
```

```python
r = requests.get(url)
```

```python
r.text
```

```python
soup = BeautifulSoup(r.content, 'lxml')
```

```python
soup
```

```python
name=[]
for node in soup.findAll('p',attrs={'class':'result-info'}):
    name.append((node.find('a').text))
    
```

```python
result=[]
for node in soup.findAll('span',attrs={'class':'result-price'}):
    result.append((node.text))
```

```python
d={}
for x,y in zip(name,result):
    d.update({x:y})
    
```
