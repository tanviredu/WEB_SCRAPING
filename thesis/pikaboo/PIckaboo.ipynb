```python
soup = BeautifulSoup(r.content, 'lxml')
```

```python
for item in soup.find_all('div',attrs={'class':'product-shop'}):
    print (item.find('h2').text)
    print (item.find('span',attrs={'class':'price','itemprop':'price'}).text)
    print ("---------------------------------")
```
